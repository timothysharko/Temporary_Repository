# spr-scala-notebook

```pyspark
import org.apache.spark.sql.SparkSession
val spark = SparkSession.builder().appName("Create View").getOrCreate()
```

```pyspark
//
// Read csv file
//
val df01 = spark.read.format("csv").option("sep", ",").option("inferSchema", "true").option("header", "true").load("s3://spr-flight-data/On_Time_Reporting_*.csv")
```

```pyspark
//
// Write parquet files
//
df01.write.parquet("flightontime.parquet")
```

```pyspark
//
// Create temporary view
//

df01.createOrReplaceTempView("flightontime")
```

```pyspark
df01.printSchema()
```

```pyspark
//
// 6,617,840 flights for 13 months
//
df01.count()
```

```pyspark
val fotcount = spark.sql("Select count(*) from flightontime")
```
