### Bank Customer Churn Modeling with Keras and TensorFlow

This program is based upon common data science programs and the Kaggle customer churn modeling dataset. I have defined my Keras neural network to be Google TensorFlow. This is a baseline program.

```python
# Import Common Libraries
import numpy as np
import numpy as np
import pandas as pd
```

```python
# Import dataset with features and targets columns
dataset = pd.read_csv('Churn_Modelling_Encoded.csv')
dataset.head()
```

```python
# Define features, target and target classes
features =[
     'CreditScore'
    ,'Geography'
    ,'Gender'
    ,'Age'
    ,'Tenure'
    ,'Balance'
    ,'NumOfProducts'
    ,'HasCrCard'
    ,'IsActiveMember'
    ,'EstimatedSalary'
]

target   = ['Exited']

classes = ['no', 'yes']

X = dataset[features]
y = dataset[target]
```

```python
# Cross validation and split the dataset into training set and test set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Perform feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Define Neural Network Model with Keras, Tensorflow Under the Covers
import keras
from keras.models import Sequential
from keras.layers import Dense

# Initialising the ANN
classifier = Sequential()

# Adding the input layer and the first hidden layer
classifier.add(Dense(output_dim = 6, init = 'uniform', activation = 'relu', input_dim = 10))

# Adding the second hidden layer
classifier.add(Dense(output_dim = 6, init = 'uniform', activation = 'relu'))

# Adding the output layer
classifier.add(Dense(output_dim = 1, init = 'uniform', activation = 'sigmoid'))

# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# Fitting the ANN to the Training set
classifier.fit(X_train, y_train, batch_size = 10, nb_epoch = 150)
```

```python
# Predicting the Test set results
y0_pred = classifier.predict(X_test)
y_pred = (y0_pred > 0.5)
```

```python
# Produce the confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```
